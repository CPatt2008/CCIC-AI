# How do the Amazon top-selling books from 2016 compare to 2010 in review number and ratings?

This project examines the top-selling books from Amazon top-selling books dating back from 2009 to now. This information was given to us by our teacher. This project is observational because I'm using information already found, and I'm not creating any new information. This information is a sample of the top books because the list given by Amazon goes up to 99. This creates a bias in the information because I cannot fully make a claim for all books, just the top 50 from every year starting in 2009. I plan to use a histogram to show the difference in the number of reviews sold from 2016 and 2010. Along with that, I will use the median, minimum, and maximum to compare these accurately. After that, I plan to use a density plot to display the difference between the ratings.

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
Allbooks = pd.read_csv("book.csv")
```

```python
Allbooks.head(10)
```

I'm going to check if the years 2016 and 2010 both have 50 or more rows of information before I use them(They do)

```python
Allbooks[Allbooks["Year"] == 2016]
```

```python
year16 = Allbooks[Allbooks["Year"] == 2016]
```

```python
Allbooks[Allbooks["Year"] == 2010]
```

```python
year10 = Allbooks[Allbooks["Year"] == 2010]
```

```python
sns.histplot(data=year16, x='Reviews', bins=10)
sns.histplot(data=year10, x='Reviews', bins=10)
#This data showcases how many reviews these books had and compares them. Books from 2016 are in the blue, while books from 2010 are orange
```

Because of two Major outliers from 2016, I will get rid of those books because they might be an error

```python
year16[year16["Reviews"] >= 40000]
#This should get both outliers
```

```python
year16 = year16.drop([383])
year16 = year16.drop([438])
#removes said outliers
```

```python
sns.histplot(data=year16, x='Reviews', bins=10)
sns.histplot(data=year10, x='Reviews', bins=10)
#just to check
```

```python
year10["Reviews"].describe()
```

```python
year16["Reviews"].describe()
```

Now, I'm going to find and compare the ratings these top 50 books had from the year 2016 and 2010 using a density plot

```python
rating16 = year16["User Rating"]
```

```python
rating10 = year10["User Rating"]
```

# Conclusion
Out of the Top 50 Amazon top-selling books in terms of reviews, the year 2016 had the highest average number of reviews with a mean of around 12,500; however, without the 2016 outliers, 2010 had the greatest maximum amount of reviews on one book, with a number over 30,000 reviews.
